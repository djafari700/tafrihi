```python
train_data['TicketPerOrder'] = train_data.groupby('BillID').TicketID.transform('count')
test_data['TicketPerOrder'] = test_data.groupby('BillID').TicketID.transform('count')
```

```python
li = train_data.groupby('BillID')['Male'].unique().index
iter = 0
all = list()

train_data['family'] = False

for i in train_data.groupby('BillID')['Male'].unique():
    if set(i) == {True, False}:
        all.append(li[iter])
    iter +=1

for i in all:
    train_data.loc[train_data['BillID'] == i, 'family'] = True
```

```python
test_data
```

```python
li = test_data.groupby('BillID')['Male'].unique().index
iter = 0
all = list()

test_data['family'] = False

for i in test_data.groupby('BillID')['Male'].unique():
    if set(i) == {True, False}:
        all.append(li[iter])
    iter +=1

for i in all:
    test_data.loc[test_data['BillID'] == i, 'family'] = True
```

```python
train_data.drop(columns=['BillID', 'TicketID', 'UserID', 'HashPassportNumber_p', 'HashEmail', 'NationalCode','VehicleType'], inplace=True)
test_data.drop(columns=['BillID', 'TicketID', 'UserID', 'HashPassportNumber_p', 'HashEmail', 'NationalCode','VehicleType'], inplace=True)
```

```python
train_data.isna().sum()
```

```python
train_data['CancelTime'].value_counts().sort_values(ascending=False).index[0]
```

```python
columns = ['VehicleClass']
for column in columns:
    most_frequent = train_data[column].value_counts().sort_values(ascending=False).index[0]
    train_data[column].fillna(most_frequent, inplace=True)
    test_data[column].fillna(most_frequent, inplace=True)
```

```python
#CONVERT TO DATETIME
columns = ['Created', 'CancelTime', 'DepartureTime']

for column in columns:
    train_data[column] = pd.to_datetime(train_data[column])
    test_data[column] = pd.to_datetime(train_data[column])
```

```python
# MOST DAYS BETWEEN CancelTime AND Created
tmp = train_data['CancelTime'] - train_data['Created'] 
tmp.dropna(inplace=True)
tmp.to_list()
day = tmp.dt.days
day.value_counts(ascending=False).index[0]
```

```python


train_data['Departure_Created'] = (train_data['DepartureTime'] - train_data['Created']).dt.days
test_data['Departure_Created'] = (test_data['DepartureTime'] - test_data['Created']).dt.days
train_data['Departure_Created']
```

```python
train_data['DepartureMonth'] = train_data['DepartureTime'].dt.month
test_data['DepartureMonth'] = test_data['DepartureTime'].dt.month
```

```python
train_data['CancelTime'].fillna(train_data['Created'],inplace =True)
test_data['CancelTime'].fillna(test_data['Created'],inplace =True)
```

```python
train_data.isna().sum()
```

```python
train_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_data['TripReason'] = le.fit_transform(train_data['TripReason'])
to_label = dict(zip(le.classes_, le.transform(le.classes_)))
to_class = dict(zip(le.transform(le.classes_), le.classes_))
print(to_label)
```

```python
test_data.drop(columns = ['Created' , 'CancelTime', 'DepartureTime'],inplace = True)
train_data.drop(columns = ['Created' , 'CancelTime', 'DepartureTime'],inplace = True)
```

```python
train_data.select_dtypes(include = "object").columns
```

```python
from sklearn.preprocessing import OrdinalEncoder

categorical_cols = train_data.select_dtypes(include = "object").columns

enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

train_data[categorical_cols] = enc.fit_transform(train_data[categorical_cols])
test_data[categorical_cols] = enc.transform(test_data[categorical_cols])
```

```python
train_data.drop(columns='BuyerMobile',inplace=True)
test_data.drop(columns='BuyerMobile',inplace=True)
```

```python
train_data = train_data.drop(train_data[train_data['CouponDiscount']<0].index)
```

```python
test_data.loc[:, 'Price'] = test_data['Price'] - test_data['CouponDiscount']
train_data.loc[:, 'Price'] = train_data['Price'] - train_data['CouponDiscount']
```

```python
train_data['Discount'] = False
test_data['Discount'] = False
```

```python
train_data.loc[train_data['CouponDiscount'] > 0,'Discount'] = True
test_data.loc[test_data['CouponDiscount'] > 0,'Discount'] = True
```

```python
train_data.drop(columns='CouponDiscount',inplace=True)
test_data.drop(columns='CouponDiscount',inplace=True)
```

```python

test_data.loc[test_data['Price'] < 0, 'Price'] *= -1

```

```python


train_data = train_data.drop(train_data[train_data['Price']<=0].index)
```

```python

# Create a DataFrame (replace this with your actual data)


# Calculate the IQR for the column
Q1 = train_data['Price'].quantile(0.25)
Q3 = train_data['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outlier removal
lower_bound = Q1 - 10 * IQR
upper_bound = Q3 + 10 * IQR

# Filter out rows with values beyond the bounds
train_data = train_data[(train_data['Price'] >= lower_bound) & (train_data['Price'] <= upper_bound)]
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming you have your train_data and test_data DataFrames
# Replace this with your actual DataFrames

# Columns to scale
columns_to_scale = ['Price']

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the 'Price' column in train_data
train_data.loc[:, columns_to_scale] = scaler.fit_transform(train_data[columns_to_scale])

# Transform the 'Price' column in test_data using the same scaler
test_data.loc[:, columns_to_scale] = scaler.transform(test_data[columns_to_scale])
```
